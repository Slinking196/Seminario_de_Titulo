```python
df_completo = pd.read_csv('data/Chile_all_csv.csv', sep=',', encoding='latin-1')
df_completo.dropna(inplace=True)
df_test = pd.read_csv('data/test.csv', sep=',', encoding='latin-1')

#print(df.head(1000))
```

```python
import re

pattern_es = re.compile("|".join([re.escape(p) for p in seed_list_es]), flags=re.IGNORECASE)

def contiene_sugerencia(texto):
    return bool(pattern_es.search(str(texto)))

df_completo["candidato_sugerencia"] = df_completo["review_text"].apply(contiene_sugerencia)
print((df_completo["candidato_sugerencia"]==True).sum())
```
